```python
# Since k means only works on numerical data, we need to convert categorical data to numerical data
# Let's check the data types of each column
df.info()
```

```python
# df['PRODUCTCODE'].unique()

# apply label encoding on product code as well
df['PRODUCTCODE'] = le_obj.fit_transform(df['PRODUCTCODE'])

df['PRODUCTCODE']
```

```python
# Now we have converted all the columns to numerical data

# jab bhi zyada correlation raha between 2 columns to model overfit hojata
# so we need to remove one of them
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))

# taking only first 10 columns

sns.heatmap(df.iloc[:, :12].corr(), annot=True)
```

```python
# jinka zyada correlation hai unko remove karna hai
# year_id and ordernumber & month_id and Qtr_id
# remove year_id and qtr_id
df.drop(['YEAR_ID','QTR_ID'], axis=1, inplace=True)
```

```python
# Now we will start K means clustering

# create a copy of df
dfk = df.copy()
```

```python
# cleaning outliers as Kmeans is sensitive to outliers
# probable outliers can be in quantity ordered, price each, sales

def clean_outliers(data):
    ninety = np.percentile(data, 90)
    ten = np.percentile(data, 10)

    data = np.where(data >= ninety, ninety, data)
    data = np.where(data <= ten, ten, data)


outlier_columns = ['QUANTITYORDERED', 'PRICEEACH', 'SALES']
for col in outlier_columns:
    clean_outliers(dfk[col])
```

```python
# Now we will do scaling on the data 
# scaling is done to bring all the data in the same range
# kmeans is based on euclidean distance, so we need to scale the data: to balance the data
# pure dataset ko scale karna hai


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dfk_scaled = scaler.fit_transform(dfk)
dfk_scaled
```

```python
# Using elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 16):
    kmeans = KMeans(n_clusters=i) # i is the number of clusters
    kmeans.fit_predict(dfk_scaled) #apply kmeans on scaled data
    wcss.append(kmeans.inertia_) # inertia_ is the sum of squared distances of samples to their closest cluster center

wcss
```

```python
# plot the elbow graph
plt.plot(range(1, 16), wcss, marker='o')    
plt.title('The Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

plt.show()
```

```python
# lets take number of clusters as 10
kmeans = KMeans(n_clusters=10, init='k-means++')
kmeans.fit_predict(dfk_scaled)
labels = kmeans.labels_

labels
```
